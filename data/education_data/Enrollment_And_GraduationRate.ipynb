```python
cursor=0
count=[[]]
for i in list_gr:
    name = pd.read_csv(i)
    if cursor<9:
        l=name[["UNITID","GRTOTLT"]]
        l=l.groupby('UNITID').agg({ 'GRTOTLT': 'sum'})
        count.append(l)
        cursor+=1

    else:
        l=name[["unitid","grrace24"]]
        l=l.groupby('unitid').agg({ 'grrace24': 'sum'})
        count.append(l)
```

```python
len(count)
first=count[1]
second=count[2]
#merged_inner = pd.merge(left=first,right= second, left_on='UNITID', right_on='UNITID')
merged_gr =pd.merge(first, second, left_index=True, right_index=True, how='inner')
for i in range(3, 17):
    third_list=count[i]
    merged_gr =pd.merge(merged_gr, third_list, left_index=True, right_index=True)
print(merged_gr)
```
