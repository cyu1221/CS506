```python
import pandas as pd
listofcsv=[
"effy2016.csv",
"effy2015.csv",
"effy2014.csv",
"effy2013.csv",
"effy2012.csv",
"effy2011.csv",
"effy2010.csv",
"effy2009.csv",
"effy2008.csv",
"effy2007.csv",
"effy2006.csv",
"effy2005.csv",
"effy2004.csv",
"effy2003.csv",
"effy2002.csv",
"ef2001d1.csv",
]
list=[[]]
len(listofcsv)
for i in range(len(listofcsv)):
    enrollment=pd.read_csv(listofcsv[i])
    if i <9:
        enrollment = enrollment[["UNITID","EFYTOTLT"]]
        enrollment=enrollment.groupby('UNITID').agg({ 'EFYTOTLT': 'sum'})
        list.append(enrollment)
    elif i==9:
        enrollment = enrollment[["UNITID","FYRACE24"]]
        enrollment=enrollment.groupby('UNITID').agg({ 'FYRACE24': 'sum'})
        list.append(enrollment)
    elif 9<i<15:
        enrollment = enrollment[["unitid","fyrace24"]]
        enrollment=enrollment.groupby('unitid').agg({ 'fyrace24': 'sum'})
        list.append(enrollment)
    else:
        enrollment = enrollment[["unitid","fyrace17"]]
        enrollment=enrollment.groupby('unitid').agg({ 'fyrace17': 'sum'})
        list.append(enrollment)
```

```python
merged_enrollment =pd.merge(list[1], list[2], left_index=True, right_index=True)
for i in range(3, 17):
    thirdlist=list[i]
    merged_enrollment =pd.merge(merged_enrollment, thirdlist, left_index=True, right_index=True)
print(merged_enrollment)
```

```python
def extractVariable(Type, Var):
    if Type=="all":
        df=merge_all
    else:
        df=merge_all.loc[merge_all['TYPE'] == Type]
    varsum=df.sum(axis=0)
    varsum=varsum[::-1]
    if Var=="en":
        return varsum[:16]
    else:
        return varsum[16:32]
    

y1=[extractVariable("for-profit", "gr"),extractVariable("for-profit", "en")]
y2=[extractVariable("not-for-profit", "gr"),extractVariable("not-for-profit", "en")]
y3=[extractVariable("public", "gr"),extractVariable("public", "all")]
y4=[extractVariable("all", "gr"),extractVariable("all", "en")]
```
