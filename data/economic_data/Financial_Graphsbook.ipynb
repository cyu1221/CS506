Log returns of above financial graphs assuming 1 year holding period
https://quantivity.wordpress.com/2011/02/21/why-log-returns/


```python
lnTNX = np.log(TNX["Adj Close"])
lnDJI = np.log(DJI["Adj Close"])
lnNSD = np.log(NSD["Adj Close"])
lnRUT = np.log(RUT["Adj Close"])
lnSNP = np.log(SNP["Adj Close"])
```

```python
def getlogreturns(df):
    buy = df.iloc[:-52].values
    sell = df.iloc[52:].values
    return sell - buy
```

```python
print(getlogreturns(lnSNP))
```

```python
mpl.plot_date(lnDJI.iloc[52:].index, getlogreturns(lnDJI), '-')
mpl.title('DJI Log Returns 1992-2018')
```

```python
mpl.plot_date(lnTNX.iloc[52:].index, getlogreturns(lnTNX), '-')
mpl.title('TNX Log Returns 1992-2018')
```

```python
mpl.plot_date(lnNSD.iloc[52:].index, getlogreturns(lnNSD), '-')
mpl.title('NASDAQ Log Returns 1992-2018')
```

```python
mpl.plot_date(lnRUT.iloc[52:].index, getlogreturns(lnRUT), '-')
mpl.title('Russell 2000 Log Returns 1992-2018')
```

```python
mpl.plot_date(lnSNP.iloc[52:].index, getlogreturns(lnSNP), '-')
mpl.title('S&P 500 Log Returns 1992-2018')
```

```python
print((getlogreturns(lnRUT)))
```

```python
logretDJI = pd.DataFrame(data = getlogreturns(lnDJI), index = lnDJI.iloc[52:].index, columns=["Log Returns"])
logretTNX = pd.DataFrame(data = getlogreturns(lnTNX), index = lnTNX.iloc[52:].index, columns=["Log Returns"])
logretNSD = pd.DataFrame(data = getlogreturns(lnNSD), index = lnNSD.iloc[52:].index, columns=["Log Returns"])
logretRUT = pd.DataFrame(data = getlogreturns(lnRUT), index = lnRUT.iloc[52:].index, columns=["Log Returns"])
logretSNP = pd.DataFrame(data = getlogreturns(lnSNP), index = lnSNP.iloc[52:].index, columns=["Log Returns"])
```

```python
logretDJI.to_csv(path_or_buf = 'DJI_log_returns.csv')
logretTNX.to_csv(path_or_buf = 'TNX_log_returns.csv')
logretNSD.to_csv(path_or_buf = 'DJI_log_returns.csv')
logretRUT.to_csv(path_or_buf = 'DJI_log_returns.csv')
logretSNP.to_csv(path_or_buf = 'DJI_log_returns.csv')
```
