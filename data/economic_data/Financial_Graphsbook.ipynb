```python
import pandas as pd
import matplotlib.pyplot as mpl

%matplotlib inline
```

```python
TNX = pd.read_csv('^TNX.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
DJI = pd.read_csv('^DJI.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
NSD = pd.read_csv('^IXIC.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
RUT = pd.read_csv('^RUT.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
SNP = pd.read_csv('^GSPC.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
#Purchasing Power of Consumer Dollar
PP = pd.read_csv('CPI_PurchPower.csv', index_col=0, parse_dates=True, infer_datetime_format=True) 
#Unemployment Rate bachelor+, 25+
Unemp = pd.read_csv('Unemployment.csv', index_col=0, parse_dates=True, infer_datetime_format=True) 
```

```python
mpl.plot_date(DJI.index, DJI['Open'], '-')
mpl.title('DJIA Open Prices 1992-2018')
```

```python
mpl.plot_date(TNX.index, TNX['Open'], '-')  #Bond yields inversely correlated to price
mpl.title('10-Year CBOE Bond Yields')
```

```python
mpl.plot_date(NSD.index, NSD['Open'], '-')
mpl.title('NASDAQ Open Prices 1992-2018')
```

```python
mpl.plot_date(RUT.index, RUT['Open'], '-')
mpl.title('Russell 2000 Open Prices 1992-2018')
```

```python
mpl.plot_date(SNP.index, SNP['Open'], '-')
mpl.title('S&P 500 Open Prices 1992-2018')
```

```python
mpl.plot_date(PP.index, PP['CUUR0000SA0R'], '-')
mpl.title('Purchasing Power of Consumer Dollar 1992-2018')
```

```python
mpl.plot_date(Unemp.index, Unemp['LNU04027662'], '-')
mpl.title('Unemployment Rate for Bachelor or Above, Age 25+ 1992-2018')
```
